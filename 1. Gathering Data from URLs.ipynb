# Gathering data from page source : text , tables , links etc 

### Text

```python
news_article = "https://www.bbc.com/news/world-europe-56465393"
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

# p paragraph
# a links
# tr : table rows
# td : table data
# for a comprehensive list of html tags : https://www.w3schools.com/tags/tag_comment.asp
mydata=urlopen(news_article).read()
```

```python
mydata
```

```python
soup = BeautifulSoup(mydata,'lxml')
```

```python
p_tags=soup.find_all('p')
```

```python
p_tags
```

```python
p_tags
```

```python
for tag in p_tags:
    print(tag.text)
```

### Links

```python
a_tags=soup.find_all('a')
```

```python
for tag in a_tags:
    link=tag.get('href')
    print(link)
    
# you can filter out links with any criterion . 
# e.g. keep only those links which begin with "https:"
```

### Tables

```python
url="https://finance.yahoo.com/quote/ADANIPOWER.NS?p=ADANIPOWER.NS&.tsrc=fin-srch"
```

```python
mydata=urlopen(url).read()
```

```python
mydata
```

```python
soup = BeautifulSoup(mydata,'lxml')
```

```python
tables=soup.find_all('table')
```

```python
tables
```

```python
mytable=tables[0]

mytable.findAll('tr')
```

```python
for rows in mytable.findAll('tr'):
    data=rows.findAll('td')
    print_data=[]
    for d in data:
        print_data.append(d.text)
    print(print_data)
```
